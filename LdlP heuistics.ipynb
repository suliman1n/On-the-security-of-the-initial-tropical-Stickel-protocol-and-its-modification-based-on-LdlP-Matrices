```python
#sym
import warnings
warnings.filterwarnings("ignore")

def sym_multi(a, b):
    value1 = max(a[0] + b[0], a[1] + b[1])
    value2 = max(a[0] + b[1], a[1] + b[0])


    return (value1, value2)


def sym_sum(a, b):
    value1 = max(a[0], b[0])
    value2 = max(a[1], b[1])

    return (value1, value2)


def sym_matrix_sum(A, B):

    n = len(A)
    m = len(A[0])


    C = [[(0, 0) for _ in range(m)] for _ in range(n)]


    for i in range(n):
        for j in range(m):
            C[i][j] = sym_sum(A[i][j], B[i][j])

    return C


def sym_matrix_multi(A, B):

    n = len(A)
    m = len(A[0])
    p = len(B[0])

    C = [[(0, 0) for _ in range(p)] for _ in range(n)]

    for i in range(n):
        for j in range(p):
            C[i][j] = (-np.inf, -np.inf)
            for k in range(m):
                C[i][j] = sym_sum(C[i][j], sym_multi(A[i][k], B[k][j]))

    return C

def Absolute(A):

    n = len(A)
    m = len(A[0])


    C = [[0 for _ in range(m)] for _ in range(n)]


    for i in range(n):
        for j in range(m):
            C[i][j] = max(A[i][j])

    return C


def symmetrize(A):
    if isinstance(A, tuple) and len(A) == 2:
        x, y = A
        if x == y:
            return (x, y)
        else:
            if x < y:
                return (-np.inf, y)
            else:
                return (x, -np.inf)

    n = len(A)
    m = len(A[0])

    C = [[(0, 0) for _ in range(m)] for _ in range(n)]

    for i in range(n):
        for j in range(m):
            x, y = A[i][j]
            if x == y:
                C[i][j] = (x, y)
            else:
                if x < y:
                    C[i][j] = (-np.inf, y)
                else:
                    C[i][j] = (x, -np.inf)

    return C




def generate_random_symmetrized_matrix(n, mm, mM):
    A = [[(random.randint(mm, mM), random.randint(mm, mM)) for _ in range(n)] for _ in range(n)]

    symmetrized_A = symmetrize(A)

    return symmetrized_A


def generate_random_symmetrized_matrix_balanced(n, mm, mM):
    A = []
    for i in range(n):
        row = []
        for j in range(n):
            if random.random() < 0.2:
                num = random.randint(mm, mM)
                row.append((num, num))
            else:
                row.append((random.randint(mm, mM), random.randint(mm, mM)))
        A.append(row)

    symmetrized_A = symmetrize(A)
    return symmetrized_A


def sym_identity(n):
    identity_matrix = [[(-np.inf, -np.inf) for _ in range(n)] for _ in range(n)]

    for i in range(n):
        identity_matrix[i][i] = (0, -np.inf)

    return identity_matrix



def sym_multi_element_to_matrix(element, matrix):
    result_matrix = []
    for row in matrix:
        result_row = []
        for item in row:
            result_row.append(sym_multi(element, item))
        result_matrix.append(result_row)
    return result_matrix
```

```python
#sym Linde
def generate_random_symmetrized_matrix_balanced_Linde(n, k, a):
    diagonal_input = random.choice([(k, k), (k, -np.inf), (-np.inf, k)])
    A = []
    for i in range(n):
        row = []
        for j in range(n):
            if random.random() < 0.2:
                num = random.randint(2*a, a+1)
                row.append((num, num))
            else:
                row.append((random.randint(2*a, a+1), random.randint(2*a, a+1)))
        A.append(row)

    for i in range(n):
        A[i][i] = diagonal_input
        
    symmetrized_A = symmetrize(A)
    return symmetrized_A





def GenerateKeyStickelsLinde(n,mm,mM,k_range,a_range):
    Wm=generate_random_symmetrized_matrix_balanced(n,mm,mM)
    k1 = np.random.randint(0, k_range)
    k2 = np.random.randint(0, k_range)
    a = np.random.randint(a_range, 1)
    b = np.random.randint(a_range, 1)
    A1 = generate_random_symmetrized_matrix_balanced_Linde(n, k1, a)
    A2 = generate_random_symmetrized_matrix_balanced_Linde(n, k2, b)

    l1 = np.random.randint(0, k_range)
    l2 = np.random.randint(0, k_range)
    c = np.random.randint(a_range, 1)
    d = np.random.randint(a_range, 1)
    B1 = generate_random_symmetrized_matrix_balanced_Linde(n, l1, c)
    B2 = generate_random_symmetrized_matrix_balanced_Linde(n, l2, d)

    temp1 = sym_matrix_multi(A1, Wm)
    U=sym_matrix_multi(temp1,A2)

    temp2 = sym_matrix_multi(B1, Wm)
    V = sym_matrix_multi(temp2, B2)

    KA = sym_matrix_multi(A1, sym_matrix_multi(V, A2))
    KB = sym_matrix_multi(B1, sym_matrix_multi(U, B2))

    if np.array_equal(symmetrize(KA), symmetrize(KB)):
        key = symmetrize(KA)
    else:
        key = []

    return key, U, V,Wm
```

```python
def find_w_st(matrix):
    max_value = -np.inf
    max_index = (-1, -1)
    max_pair = None

    for i, row in enumerate(matrix):
        for j, pair in enumerate(row):
            a, b = pair
            if a == -np.inf and b != -np.inf:
                finite_number = b
            elif b == -np.inf and a != -np.inf:
                finite_number = a
            else:
                continue

            if finite_number > max_value:
                max_value = finite_number
                max_index = (i, j)
                max_pair = pair

    return max_pair, max_index


def convert_to_w_st_(pair):
    return (-(pair[0]) if pair[0] != -np.inf else pair[0], 
            -(pair[1]) if pair[1] != -np.inf else pair[1])
```

```python
def Attack_1(Wm,U,V):
    U=symmetrize(U)
    V=symmetrize(V)
    max_value = -np.inf
    max_index = (-1, -1)
    max_pair = None

    for i, row in enumerate(Wm):
        for j, pair in enumerate(row):
            a, b = pair
            if a == -np.inf and b != -np.inf:
                finite_number = b
            elif b == -np.inf and a != -np.inf:
                finite_number = a
            else:
                continue

            if finite_number > max_value:
                max_value = finite_number
                max_index = (i, j)
                max_pair = pair

    w_st=max_pair
    s,t=max_index
    
    k1_k2=sym_multi(U[s][t],convert_to_w_st_(w_st))
    l1_l2=sym_multi(V[s][t],convert_to_w_st_(w_st))
    
    
    key_attack=sym_matrix_sum(sym_multi_element_to_matrix(l1_l2, U),sym_multi_element_to_matrix(k1_k2, V))
    
    return symmetrize(key_attack)
```

```python
def Attack_2(Wm,U,V):
    n=len(U)
    U=symmetrize(U)
    V=symmetrize(V)
    max_value = -np.inf
    max_index = (-1, -1)
    max_pair = None

    for i, row in enumerate(Wm):
        for j, pair in enumerate(row):
            a, b = pair
            if a == -np.inf and b != -np.inf:
                finite_number = b
            elif b == -np.inf and a != -np.inf:
                finite_number = a
            else:
                continue

            if finite_number > max_value:
                max_value = finite_number
                max_index = (i, j)
                max_pair = pair

    w_st=max_pair
    s,t=max_index
    
    k1_k2=sym_multi(U[s][t],convert_to_w_st_(w_st))
    l1_l2=sym_multi(V[s][t],convert_to_w_st_(w_st))
    
    
    key_attack=sym_matrix_sum(sym_multi_element_to_matrix(l1_l2, U),sym_multi_element_to_matrix(k1_k2, V))
    
    key_attack = [[(0, 0) for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            te1=sym_multi(V[s][t],U[i][j])
            te2=sym_multi(U[s][t],V[i][j])
            te3=sym_multi(U[i][t],V[s][j])
            te4=sym_multi(V[i][t],U[s][j])
            
            re1=sym_sum(te1,te2)
            re2=sym_sum(te3,te4)
            re=sym_sum(re1,re2)
            
            key_attack[i][j] =sym_multi(convert_to_w_st_(w_st),re)
    
    return symmetrize(key_attack)
```
